```python
class empleados():

    empresa = "Data Analytics"   #atributos de clase
    ascenso = 1.5
    n_subido_sueldo = 0

    def __init__(self, nombre, apellido, salario, clave,*args):  #atributos de instancia
         self.nombre = nombre
         self.apellido = apellido
         self.salario = salario
         self.__clave = clave  #esto hace que cuando intentemos llamarla python sepa que es una variable que no sea accesible por el usuario
         self.args = args
       
    
    def ascender(self):  #metodo de instancia y lo llamamos cuando referimos al objeto empleado2
        self.salario = self.salario*self.ascenso
        print(f"Hola, {self.nombre}. Has sido ascendida. Tu nuevo salario es de {self.salario: .0f} euros")
        empleados.n_subido_sueldo = self.n_subido_sueldo + 1

    def cambiar_nombre_emp():    #metodo de clase, solo hace referencia a la clase, aqui llamo a empleados
        empleados.empresa = "Data Analytics S.L"

    def mostrar_clave(self):  #metodo de instancia dentro de clase para mostrar la clave, es la unica manera de que la muestre porque pusimos __
        print(f' la clave es {self.__clave}')
     

  


```
